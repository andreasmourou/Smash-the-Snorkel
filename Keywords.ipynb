```python
import pandas as pd
import string
import numpy as np
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
stop = stopwords.words('english')
import collections
from collections import Counter
```

```python
#import development set, remove unessesaty columns, rename
df=pd.read_csv("/project/review_dev_labelled.csv")
df=df.reset_index()
df=df.drop(["index_1", "index_2","label"], axis=1)
df=df.rename(columns={"Label 1 if text contains a suggestion/complaint": "label"})
```

```python
#remoce punctuation from the string 
df["text"]=df['text'].str.replace('[{}]'.format(string.punctuation), '')
#split to 2 different dataframes 
one=df[df["label"]==1]
zero=df[df["label"]==0]
```

```python
#remove stop words and split the string into individual words from the dateframe with label 1
one['text_new'] = one['text'].str.lower().apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
b=one["text_new"].str.split(expand=True).stack().value_counts()
b=b.reset_index()
b.rename(columns={ b.columns[1]: "count" }, inplace = True)
# pd.options.display.max_rows = 999
# b.sort_values(by="count", ascending=False).head(200)
```

```python
##remove stop words and split the string into words from the dateframe with label=0

zero['text_new'] = zero['text'].str.lower().apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
a=zero["text_new"].str.split(expand=True).stack().value_counts()
a=a.reset_index()
a.rename(columns={ a.columns[1]: "count" }, inplace = True)
# pd.options.display.max_rows = 9999
# a.sort_values(by="count", ascending=False).head(200)
```

```python
# b.rename(columns={ b.columns[1]: "count" }, inplace = True)
```

```python
#Convert the string column to list
list_one=b["index"].values.tolist()
list_zero=a["index"].values.tolist()
# Find the words that appear in the reviews of label=1 but not in label=0
Difference_one_zero=list(set(list_one) - set(list_zero))
# Find the words that appear in the reviews of label=0 but not in label =1
Difference_zero_one=list(set(list_zero) - set(list_one))
```

```python
x=pd.DataFrame(Difference_one_zero)
x.rename(columns={ x.columns[0]: "count" }, inplace = True)
y=pd.DataFrame(Difference_zero_one)
y.rename(columns={ y.columns[0]: "count" }, inplace = True)
```

```python
#Print all the unique words that are in label=1 but not in label=0
unique_word_for_complaints=x["count"].values.tolist()
for p in unique_word_for_complaints:
    print(p)
```

```python
#Print all the unique words that are in label=0 but not in label=1
unique_words_for_zero=y["count"].values.tolist()
for p in unique_words_for_zero:
    print(p)
```
