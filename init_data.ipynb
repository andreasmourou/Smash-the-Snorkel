```python
import json 
import pandas as pd 
from pandas.io.json import json_normalize 
```

```python
df_biz = pd.read_json('./original_data/business.json', lines=True)
```

```python
df_biz_sample = df_biz.sample(random_state=42, n=30 )
```

```python
df_biz_sample.head()
```

```python
df_check = pd.read_json('./original_data/checkin.json', lines=True)
df_check.head()
```

```python
df_check_sample = df_check[df_check['business_id'].isin(df_biz_sample['business_id'])]
```

```python
# df_review = pd.read_json('review.json', lines=True)
# df_review.head()
df_review = pd.DataFrame()
df_reader = pd.read_json('review.json', lines=True, chunksize=100000)
for chunk in df_reader:
    df_review=df_review.append(chunk)
```

```python
df_review.head()
```

```python
df_review_sample = df_review[df_review['business_id'].isin(df_biz_sample['business_id'])]
```

```python
df_tip = pd.DataFrame()
df_reader = pd.read_json('tip.json', lines=True, chunksize=100000)
for chunk in df_reader:
    df_tip=df_tip.append(chunk)
```

```python
df_tip_sample=  df_tip[df_tip['business_id'].isin(df_biz_sample['business_id'])]
```

```python
df_user = pd.DataFrame()
df_reader = pd.read_json('user.json', lines=True, chunksize=100000)
for chunk in df_reader:
    df_user=df_user.append(chunk)
```

```python
df_user_sample=  df_user[df_user['user_id'].isin(df_review_sample['user_id']) & df_user['user_id'].isin(df_tip_sample['user_id'])]
```

```python
exportlist=[df_biz_sample,df_check_sample,df_review_sample,df_tip_sample]
namelist=['df_biz_sample','df_check_sample','df_review_sample','df_tip_sample']
for i in range(4):
    exportlist[i].to_csv(namelist[i]+'.csv',index=False)
```
